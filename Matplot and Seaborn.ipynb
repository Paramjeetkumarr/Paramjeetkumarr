## Sine Curve 

```python
# np.arange(0,4*np.pi(Random Between),0.1(step-size))
x= np.arange(0,4*np.pi,0.1)
```

```python
x = np.arange(0,5*np.pi,0.1)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.title('Sin wave')
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')

plt.subplot(2,2,2)
plt.plot(x,y_cos)
plt.title('Cos Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

## Box Plot 

Syntax-:
plt.boxplot(data, vert=True(**Vertical**), patch_artist = True(**Color**))

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(data, vert=True, patch_artist = True)

plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Box plot with outlier')
```

## Pie Chart
Syntax:-
plt.pie(size,explode= explods(**Extract Piece**), labels=label(**Label of data**), colors= Color(**Color Coding**),autopct='%1.1f%%'(**% of data in 100% grand total**), shadow = True)


```python
label = 'Python', 'C++', 'Ruby', 'Java'
size = [215,139,245,210]
Color = ['Gold','yellowgreen','Lightcoral','lightskyblue']
explods = (0.1,0,0,0)
```

```python
plt.pie(size,explode= explods, labels=label, colors= Color,autopct='%1.1f%%', shadow = True)

plt.title('Pie Chart of programming lang')
```

```python
# Simple Pie Chart
label = 'Python', 'C++', 'Ruby', 'Java'
size = [215,139,245,210]
plt.pie(size, labels= label)

plt.title("Pie Chart")
```

## Correlation with Heatmap 

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important in feature selection

Syntax-:
 Sns.heatmap(df.corr())

```python
#Only Integer and float
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint Plot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It usually a scatter plot, a hexbin plot, a 2d histogram or a 2d density plot

Syntax-:
sns.jointplot(x='xlabel',y='ylabel', data=data, kind='hex'/'reg')

### Univariate Analysis 

```python
#Comparing Tip with Total Bill

sns.jointplot(x='tip',y='total_bill', data=df, kind='hex')
```

```python
#kind='reg' - Show all point as regeression
# Best fitted line- Reg line
# PDF - Probability Density function

sns.jointplot(x='tip',y='total_bill', data=df, kind='reg')
```

## Count Plot

It shows the count of observation in each categorical bin using bars.

Syntax -:
sns.countplot(y(horizontal)='Categorical_Column', data = dataframe)

```python
sns.countplot('sex', data=df)
```

```python
#x='Sex'-(By default vertical)
#y='Sex'- (Horizontal)

sns.countplot(y='sex', data = df)
```

## Box plot 

A box and wishker plot is a graph that represent information from a five number summary

Syntax-:
sns.boxplot(x='xlabel',y='ylabel',data =dataframe, palette(Color)='rainbow', orient='h'(Horizontal))

```python
sns.boxplot('sex','tip', data= df)
```

Violin plot helps to see both the distribution of the data in terms of kernel density estimation and the box plot.
